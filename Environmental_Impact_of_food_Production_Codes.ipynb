### Handle Missing Data

```python
# Drop columns with excessive missingness
threshold = len(df) * 0.5
df_clean = df.dropna(axis=1, thresh=threshold)

# Impute remaining numerical columns with median
numerical_cols = df_clean.select_dtypes(include=np.number).columns
df_clean[numerical_cols] = df_clean[numerical_cols].fillna(df_clean[numerical_cols].median())
```

### Exploratory Data Analysis (EDA)

```python
# Check if Total Impact Score exists
if "Total Impact Score" in df.columns:
    print("Total Impact Score exists:", df["Total Impact Score"].head())
else:
    print(" 'Total Impact Score' is missing!")

# Ensure skewed_cols is defined before using it
skewed_cols = [
    "Total_emissions", 
    "Land use per kilogram (m² per kilogram)", 
    "Scarcity-weighted water use per kilogram (liters per kilogram)", 
    "Eutrophying emissions per kilogram (gPO₄eq per kilogram)"
]

# Check if all required columns exist
missing_cols = [col for col in skewed_cols if col not in df.columns]
if missing_cols:
    print(f" Missing columns: {missing_cols}")

# Check for NaN or negative values before log transformation
for col in skewed_cols:
    print(f"{col}: {df[col].isna().sum()} NaN values, {(df[col] < 0).sum()} negative values")

# Recreate Total Impact Score if needed
if "Total Impact Score" not in df.columns:
    df[skewed_cols] = df[skewed_cols].fillna(0).apply(lambda x: np.log1p(x))  # Handle NaN
    df["Total Impact Score"] = df[skewed_cols].sum(axis=1)
    print(" 'Total Impact Score' has been recreated.")
```

```python
df["Total Impact Score"] = df[skewed_cols].sum(axis=1)
```

```python
# Categorize impact

def categorize_impact(score):
    if score <= 10:
        return "Low Impact"
    elif 10 < score <= 20:
        return "Medium Impact"
    else:
        return "High Impact"

df["Impact Category"] = df["Total Impact Score"].apply(categorize_impact)
```

#### Insight 1: Beef and Lamb Have the Highest Emissions

```python
# Highlight Beef and Lamb products
plt.figure(figsize=(12, 6))
top_10 = df_clean.nlargest(10, 'Total_emissions')

# Custom colors for Beef/Lamb vs others
colors = ['firebrick' if ('Beef' in food or 'Lamb' in food) else 'teal' for food in top_10['Food product']]

sns.barplot(
    x='Total_emissions', 
    y='Food product', 
    data=top_10, 
    palette=colors
)
plt.title('Top 10 High-Emission Foods (Highlighting Beef/Lamb)', fontsize=14)
plt.xlabel('Total Emissions (kg CO₂eq)')
plt.ylabel('')
plt.show()

# Statistical summary for Beef/Lamb vs others
beef_lamb = df_clean[df_clean['Food product'].str.contains('Beef|Lamb')]
others = df_clean[~df_clean['Food product'].str.contains('Beef|Lamb')]

print("Average emissions for Beef/Lamb:", beef_lamb['Total_emissions'].mean())
print("Average emissions for other foods:", others['Total_emissions'].mean())
```

### Beef and lamb have the highest emissions. Recommend reducing consumption.
#### Finding:
Beef (both beef and dairy herds) and lamb/mutton are the highest-emitting foods, with average emissions of 59.6 kg CO₂eq/kg and 24.5 kg CO₂eq/kg, respectively. These emissions are 6–15 times higher than plant-based foods like wheat (1.4 kg CO₂eq/kg) or potatoes (0.3 kg CO₂eq/kg).

#### Significance:
Animal-based products, particularly ruminants like cows and sheep, require extensive land for grazing, produce methane (a potent greenhouse gas), and involve energy-intensive feed production. This makes them disproportionately harmful to the climate compared to other foods.

#### Recommendation:

1. Reduce consumption of beef and lamb through public awareness campaigns.

2. Promote alternatives like plant-based proteins (e.g., legumes, tofu) or lower-emission meats (e.g., poultry).

#### Finding:
The "Farm" stage (e.g., fertilizer use, machinery, livestock digestion) and "Land use change" (e.g., deforestation for agriculture) contribute ~80% of total emissions across most products. For example, beef production emits 39.4 kg CO₂eq/kg at the farm stage alone.

#### Significance:
Unsustainable farming practices (e.g., synthetic fertilizers, overgrazing) and deforestation accelerate climate change and biodiversity loss.

#### Recommendation:

1. Adopt regenerative farming: Use crop rotation, organic fertilizers, and agroforestry to reduce emissions.

2. Protect forests: Enforce policies against deforestation for agriculture.

#### Insight 3: Plant-Based Foods Have Lower Emissions

```python
from scipy.stats import ttest_ind

# Boxplot comparison
plt.figure(figsize=(8, 5))
sns.boxplot(
    x='Food Type', 
    y='Total_emissions', 
    data=df_clean, 
    palette='Set2',
    showfliers=False  # Hide outliers for clarity
)
plt.title('Emissions: Plant vs. Animal-Based Foods', fontsize=14)
plt.ylabel('Total Emissions (kg CO₂eq)')
plt.show()

# T-test for statistical significance
plant_emissions = df_clean[df_clean['Food Type'] == 'Plant']['Total_emissions']
animal_emissions = df_clean[df_clean['Food Type'] == 'Animal']['Total_emissions']

t_stat, p_value = ttest_ind(plant_emissions, animal_emissions, nan_policy='omit')
print(f"T-test p-value: {p_value:.4f}")

# Mean comparison
print("\nMean Emissions:")
print(f"Plant-based: {plant_emissions.mean():.2f} kg CO₂eq")
print(f"Animal-based: {animal_emissions.mean():.2f} kg CO₂eq")
```

### Plant-based foods have lower emissions. Promote plant-based diets.
#### Finding:
Plant-based foods emit 5–10 times less CO₂eq than animal-based foods. For example:

#### Statistical Validation:

#### Recommendation:

1. Shift diets: Encourage plant-rich diets through public health campaigns.
